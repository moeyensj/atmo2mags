## Test 5: Change deltaGrey range

```python
comp1best, comp2best, dgbest, dmagsbest, logL, chisquared, chisquaredbest = ab.computeAtmoFit('O2','Rayleigh',atmo_obs,deltaGrey=20.0,componentBins=25,deltaGreyBins=25,pickleString='exp4',dmagLimit=False,returnData=True, deltaGreyRange=[-10.,10.])
```

## Test 6: Changing best fit components

```python
bestFit1 = 0.5
bestFit2 = 4.5

newFit1 = {}
newFit2 = {}

for i,j in enumerate(ab.filterlist):
    newFit1[j] = bestFit1
    newFit2[j] = bestFit2
```

```python
ab.regressionPlot('O2', newFit1, 'Rayleigh', newFit2, dgbest, logL, atmo_obs, componentBins=25, deltaGrey=20.0, deltaGreyBins=51)
```
