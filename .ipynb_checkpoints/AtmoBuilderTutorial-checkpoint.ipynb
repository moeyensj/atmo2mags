#### Installation:

#### Importing and initializing:

```python
from AtmoBuilder import AtmoBuilder

# shows plots within notebook cells
% matplotlib inline 
```

```python
# Initialize the class
# By default AtmoBuilder will look in the current directory for a 'modtran/' folder containing the relevant 
# MODTRAN files and read them in. It will also read in LSST filter and hardware data from the LSST stack
ab = AtmoBuilder()
```

#### Creating a sample atmosphere:

```python
atmo_test = ab.buildAtmo([1.2,0.9,1.9,2.1,0.8,1.6],1.4)
```

This creates an Atmo object (a bandpass object with additional attributes).

```python
print atmo_test.P, atmo_test.X
```
