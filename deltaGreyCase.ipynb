```python
from AtmoBuilder import AtmoBuilder

import numpy as np
import matplotlib.pyplot as plt
import copy

% matplotlib inline
```

```python
ab = AtmoBuilder()
```

```python
ab.readAll()
```

```python
atmo_obs = ab.buildAtmo([1.0,1.0,0.5,1.0,1.0,1.7],2.0)
atmo_std = ab.buildAtmo(ab.parameters, 1.2)
throughput_obs = ab.combineThroughputs(atmo_obs)
throughput_std = ab.combineThroughputs(atmo_std)
```

```python
ab.computeDeltaGreyFit('O3',12.0,atmo_obs,deltaGreyRange=[-20,20],deltaGreyBins=50)
```

```python
compbest, dgbest, dmagsbest, logL, chisquared, chisquaredbest, dmags_obs = ab.computeDeltaGreyFit('O3',12.0,atmo_obs,deltaGreyRange=[-20,20],deltaGreyBins=51,returnData=True)
```

```python
def bestFitOverride(comp, compbest, newcompbest, deltaGrey, dgbest, newdgbest, logL, filters=ab.filterlist):
    compbest_new = copy.deepcopy(compbest)
    dgbest_new = copy.deepcopy(dgbest)
    throughput_new = copy.deepcopy(throughput_obs)
    
    P_fit = copy.deepcopy(atmo_obs.P)
    X_fit = copy.deepcopy(atmo_obs.X)
    
    range1, P_num = ab._componentCheck(comp, 20)
    P_fit[P_num] = newcompbest
    atmo_fit = ab.buildAtmo(P_fit,X_fit)
    
    for f in filters:
        compbest_new[f] = newcompbest
        dgbest_new[f] = newdgbest
        throughput_new[f] = ab.combineThroughputs(atmo_fit,filters=f)[f]    
       
    ab.dphiPlot(throughput_obs, throughput_std, bpDict2=throughput_new, filters=ab.filterlist, regression=True)
    ab.ddphiPlot(throughput_obs, throughput_new, throughput_std, filters=ab.filterlist, regression=True,)    
    ab.regressionPlotDeltaGrey(comp, compbest_new, deltaGrey, dgbest_new, logL, atmo_obs, deltaGreyRange=[-20,20],deltaGreyBins=51)
    
    return
```

```python
bestFitOverride('O3', compbest, 1.0, 12.0, dgbest, 7.0, logL, filters=['u'])
```

```python
bestFitOverride('O3', compbest, 2.0, 12.0, dgbest, 0.0, logL, filters=['u'])
```

```python
bestFitOverride('O3', compbest, 3.0, 12.0, dgbest, -7.0, logL, filters=['u'])
```

```python
bestFitOverride('O3', compbest, 3.0, 12.0, dgbest, 10.0, logL, filters=['r'])
```
