```python
import copy

P_fit = copy.deepcopy(atmo_obs.P) 
X_fit = copy.deepcopy(atmo_obs.X)
throughput_fit = {}

for f in ab.filterlist:
    P_fit[0] = compbest[f]
    print atmo_fit_u.P
    atmo_fit = ab.buildAtmo(P_fit,X_fit)
    throughput_fit[f] = ab.combineThroughputs(atmo_fit, filters=f)[f]
    print max(throughput_fit[f].phi), min(throughput_fit[f].phi), np.median(throughput_fit[f].phi)

ab.dphiPlot(tp_obs, tp_std, bpDict2=throughput_fit, regression=True)
ab.ddphiPlot(tp_obs, throughput_fit, tp_std, regression=True)
```
