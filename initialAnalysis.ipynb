```python
from AtmoCompMod import AtmoComp as ac
from plot_dmagsMod import plot_dmags as dm

import os
import numpy
import pylab
import matplotlib

% matplotlib inline
```

```python
# The standard atmosphere using modtran files
stdx10 = open('Pachon_MODTRAN.10.7sc','r')
components = ('H2O','O2','O3','rayleigh','aerosol')

def aero(x,w,a):
    return numpy.e**-(0.1*x*(550.0/w)**a)

wave = []
trans = {}

for comp in components:
    trans[comp] = []

for lineNum,line in enumerate(stdx10):
    lineEle = line.split()
    if lineNum<4:
        continue
    if float(lineEle[0]) < 0.0:
        break  
    wave.append(float(lineEle[0]))
    trans['H2O'].append(float(lineEle[2]))
    trans['O2'].append(float(lineEle[3]))
    trans['O3'].append(float(lineEle[4]))
    trans['rayleigh'].append(float(lineEle[8]))
    trans['aerosol'].append(aero(1.0,float(lineEle[0]),1.7))

wavelen = numpy.array(wave,dtype='float')
trans['H2O'] = numpy.array(trans['H2O'],dtype='float')
trans['O2'] = numpy.array(trans['O2'],dtype='float')
trans['O3'] = numpy.array(trans['O3'],dtype='float')
trans['rayleigh'] = numpy.array(trans['rayleigh'],dtype='float')
trans['aerosol'] = numpy.array(trans['aerosol'],dtype='float')
```

```python
comb = trans['H2O']*trans['O2']*trans['O3']*trans['rayleigh']*trans['aerosol']

fig,ax = pylab.subplots(1,1)
fig.set_size_inches(10,7)

ax.plot(wavelen,comb)
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Transmission')
ax.set_title('Standard Atmosphere')
```

```python
w = numpy.arange(250,1100.5,0.5,dtype='float')

def aero(x,w,a):
    return numpy.e**-(0.1*x*(550.0/w)**a)

fig, ax = pylab.subplots(1,2)
fig.set_size_inches(14,7)
fig.tight_layout()

ax[0].plot(w,aero(1.0,w,1.7),label='X: 1.0')
ax[0].plot(w,aero(1.2,w,1.7),label='X: 1.2')
ax[0].plot(w,aero(1.4,w,1.7),label='X: 1.4')
ax[0].plot(w,aero(1.6,w,1.7),label='X: 1.6')
ax[0].plot(w,aero(1.8,w,1.7),label='X: 1.8')
ax[0].plot(w,aero(2.0,w,1.7),label='X: 2.0')
ax[0].set_xlabel('Wavelength (nm)')
ax[0].set_ylabel('Transmission')
ax[0].set_title('Aerosol Transmission vs. Wavelength at Varying Airmasses (alpha: 1.7)')
ax[0].legend(loc='lower right',shadow='False')

ax[1].plot(w,aero(1.0,w,2.0),label='X: 1.0')
ax[1].plot(w,aero(1.2,w,2.0),label='X: 1.2')
ax[1].plot(w,aero(1.4,w,2.0),label='X: 1.4')
ax[1].plot(w,aero(1.6,w,2.0),label='X: 1.6')
ax[1].plot(w,aero(1.8,w,2.0),label='X: 1.8')
ax[1].plot(w,aero(2.0,w,2.0),label='X: 2.0')
ax[1].set_xlabel('Wavelength (nm)')
ax[1].set_ylabel('Transmission')
ax[1].set_title('Aerosol Transmission vs. Wavelength at Varying Airmasses (alpha: 2.0)')
ax[1].legend(loc='lower right',shadow='False')
```

```python
def atmo(w,h2o,o2,o3,rayleigh,aerosol):
    comb = numpy.e**(h2o*trans['H2O']+o2*trans['O2']+o3*trans['O3']+rayleigh*trans['rayleigh']+aerosol*trans['aerosol'])
    return comb
```
