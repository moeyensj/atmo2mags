```python
from AtmoBuilder import AtmoBuilder

import numpy as np
import matplotlib.pyplot as plt

% matplotlib inline
```

```python
ab = AtmoBuilder()
```

```python
ab.readAll()
```

```python
atmo_obs = ab.buildAtmo([1.0,1.0,1.0,2.0,1.3,1.7],1.4)
tp_obs = ab.combineThroughputs(atmo_obs)

atmo_std = ab.buildAtmo(ab.parameters,1.2)
tp_std = ab.combineThroughputs(atmo_std)
```

```python
ab.dmagPlot(tp_obs,tp_std,'qsos')
```

```python
ab.dmagPlot(tp_obs,tp_std,'gals',deltaGrey=200.0)
```
